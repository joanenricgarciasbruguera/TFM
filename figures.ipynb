### Susceptibilitat

#### FIGURA 3

```python
dfX_T3 = pd.read_csv("calculs/X_vs_N_z4_T3.csv")
dfX_T4 = pd.read_csv("calculs/X_vs_N_z4_T4.csv")
dfX_T5 = pd.read_csv("calculs/X_vs_N_z4_T5.csv")
dfX_T6 = pd.read_csv("calculs/X_vs_N_z4_T6.csv")
```

**T=3**

```python
log_N3 = np.log10(np.array(dfX_T3["N"]))
log_N3 = np.reshape(log_N3,(-1,1))

log_X3 = np.log10(np.array(dfX_T3["X"]))
log_X3 = np.reshape(log_X3,(-1,1))

reg3 = LinearRegression().fit(log_N3, log_X3)
log_X3_pred = (reg3.coef_[0][0] * log_N3) + reg3.intercept_
print("alpha = ", reg3.coef_[0][0])
print("R2 = ", reg3.score(log_N3, log_X3))
```

**T=4**

```python
log_N4 = np.log10(np.array(dfX_T4["N"]))
log_N4 = np.reshape(log_N4,(-1,1))

log_X4 = np.log10(np.array(dfX_T4["X"]))
log_X4 = np.reshape(log_X4,(-1,1))

reg4 = LinearRegression().fit(log_N4, log_X4)
log_X4_pred = (reg4.coef_[0][0] * log_N4) + reg4.intercept_
print("alpha = ", reg4.coef_[0][0])
print("R2 = ", reg4.score(log_N4, log_X4))
```

**T=5**

```python
log_N5 = np.log10(np.array(dfX_T5["N"]))
log_N5 = np.reshape(log_N5,(-1,1))

log_X5 = np.log10(np.array(dfX_T5["X"]))
log_X5 = np.reshape(log_X5,(-1,1))

reg5 = LinearRegression().fit(log_N5, log_X5)
log_X5_pred = (reg5.coef_[0][0] * log_N5) + reg5.intercept_
print("alpha = ", reg5.coef_[0][0])
print("R2 = ", reg5.score(log_N5, log_X5))
```

**T=6**

```python
log_N6 = np.log10(np.array(dfX_T6["N"]))
log_N6 = np.reshape(log_N6,(-1,1))

log_X6 = np.log10(np.array(dfX_T6["X"]))
log_X6 = np.reshape(log_X6,(-1,1))

reg6 = LinearRegression().fit(log_N6, log_X6)
log_X6_pred = (reg6.coef_[0][0] * log_N6) + reg6.intercept_
print("alpha = ", reg6.coef_[0][0])
print("R2 = ", reg6.score(log_N6, log_X6))
```

```python
plt.figure(figsize=[8,6])

plt.scatter(np.log10(dfX_T3["N"]), np.log10(dfX_T3["X"]), 
            s= 20, marker="o", color = "red", label = r"$T=3$")
plt.plot(log_N3, log_X3_pred, 
         linewidth=0.6, linestyle ="--", color = "red")

plt.scatter(np.log10(dfX_T4["N"]), np.log10(dfX_T4["X"]), 
            s= 20, marker="o", color = "limegreen", label = r"$T=4$")
plt.plot(log_N4, log_X4_pred,
         linewidth=0.6, linestyle ="--", color = "limegreen")

plt.scatter(np.log10(dfX_T5["N"]), np.log10(dfX_T5["X"]), 
            s= 20, marker="o", color = "blue", label = r"$T=5$")
plt.plot(log_N5, log_X5_pred, 
         linewidth=0.6, linestyle ="--", color = "blue")

plt.scatter(np.log10(dfX_T6["N"]), np.log10(dfX_T6["X"]), 
            s= 20, marker="o", color = "magenta", label = r"$T=6$")
plt.plot(log_N6, log_X6_pred,
         linewidth=0.6, linestyle ="--", color = "magenta")


plt.xlabel(r"$log_{10}(N)$", fontsize=14)
plt.ylabel(r"$log_{10}(\chi)$", fontsize=14)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11, frameon=False)


```

#### FIGURA 11

$ \displaystyle \bar{L} = \frac{1}{t_{f}-t_{i}}  \sum_{t=t_{i}}^{t_{f}} L_t $

```python
dfLmean = pd.read_csv("calculs/Lmean_eps.csv")
```

```python
log_eps = np.log10(np.array(dfLmean["eps"])[0:11])
log_eps = np.reshape(log_eps,(-1,1))

log_L = np.log10(np.array(dfLmean["L"])[0:11])
log_L = np.reshape(log_L,(-1,1))

reg = LinearRegression().fit(log_eps, log_L)
log_L_pred = (reg.coef_[0][0] * np.log10(np.array(dfLmean["eps"]))) + reg.intercept_
print("gamma = ", reg.coef_[0][0])
print("R2 = ", reg.score(log_eps, log_L))
```

```python
plt.figure(figsize=[8,6])
plt.scatter(np.log10(dfLmean["eps"]), np.log10(dfLmean["L"]), s=20,
            marker="o", color="black", label = r"$z=4, \, N=10^4$")
plt.plot(np.log10(np.array(dfLmean["eps"])), log_L_pred,
            color="black", linewidth=0.6, linestyle="--", 
             label=r"$\bar{L} \sim \epsilon^{\gamma}$")


plt.xlabel(r"$log_{10}(\epsilon)$", fontsize=14)
plt.ylabel(r"$log_{10}(\bar{L})$", fontsize=14)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(loc=4, fontsize=11, frameon=False)
```

**FIG 11**. Dependència de la mida mitjana del nucli dels agents $\bar{L}$ de $\varepsilon$ (escala log-log) al model amb reconnexió antipreferencial per $z=4$ i $N=10^{4}$. La llei de potències $\bar{L} \sim \varepsilon^{\gamma}$ és vàlida per $\varepsilon<0.1$, on $\gamma \approx 0.5947$ per $z=4$ i $N=10^{4}$. $\bar{L}$ s'ha calculat fent la mitjana temporal de $L$ dins l'interval $t \in [5 \times 10^{3}, 10^{4}]$.
